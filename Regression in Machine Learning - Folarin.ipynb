```python
df.drop(columns = ["date", "lights"], inplace =True )
```

```python
df.head()
```

```python
scaler = MinMaxScaler()

new_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
X= new_df.drop(columns = ["Appliances"])
y = new_df['Appliances']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
linear = LinearRegression()
linear.fit(X_train, y_train)
```

```python
y_pred = linear.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
rss = np.sum(np.square(y_test - y_pred))
rss
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
lmw = pd.Series(linear.coef_, X_train.columns).sort_values()
lmw = pd.DataFrame(lmw)
lmw = lmw.reset_index()
lmw.rename(columns = {"index": "Features", 0:"Linear Regression"}, inplace=True)
lmw
```

```python
lmw[lmw["Linear Regression"] == lmw["Linear Regression"].min()]
```

```python
ridge = Ridge(alpha = 0.4)
```

```python
ridge.fit(X_train, y_train)
new_p = ridge.predict(X_test)
```

```python
n_rmse = np.sqrt(mean_squared_error(y_test, new_p))
n_rmse
```

```python
las = Lasso(alpha = 0.001)
las.fit(X_train, y_train)
las_p = las.predict(X_test)
```

```python
l_rmse = np.sqrt(mean_squared_error(y_test, las_p))
l_rmse
```

```python
mw = pd.Series(las.coef_, X_train.columns).sort_values()
mw = pd.DataFrame(mw)
mw = mw.reset_index()
mw.rename(columns = {"index": "Features", 0:"Lasso"}, inplace=True)
mw
```
